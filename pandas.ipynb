For example, [select-dtypes documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.select_dtypes.html#pandas-dataframe-select-dtypes)

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('car_crashes')
```

```python
df.shape
```

```python
df.head()
```

```python
df_sub = df.select_dtypes(include = np.number).copy()
df_sub.head()
```

```python
df[df_sub < 10] = 3
```

```python
df_sub[df_sub < 10] = 3
```

```python
df_sub.head()
```

```python
df[df_sub.columns] = df_sub
```

```python
chart = alt.Chart(df_driving).mark_line().encode(
    x = alt.X("miles",scale=alt.Scale(zero=False),sort=None),
    y = alt.Y("gas",scale=alt.Scale(zero=False)),
    order = "year",
    tooltip = "year",
).properties(
    width=1000,
)
```

```python
chart + chart.mark_text(align = 'left', dx = 10).encode(
    x = 'miles',
    y = alt.value(10),
    text = "year",
)
```

```python
chart+alt.Chart(df_driving[::5]).mark_text(align= 'left', dx = 10,dy = -10).encode(
    x = 'miles',
    y = 'gas',
    text = 'year'
)
```
